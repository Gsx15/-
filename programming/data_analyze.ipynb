```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
# 去除警告框
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
# 读取数据
data_train = pd.read_csv("../titanic/train.csv")
data_test = pd.read_csv("../titanic/test.csv")
```

```python
'''
    通过matplotlib画出各个字段和是否幸存之间的关系
'''

# 解决中文乱码问题
plt.rcParams['font.sans-serif']=['SimHei']


# 定义一个显示框
plt.figure('Field_survived', facecolor='white', figsize=(10, 50), dpi=500)


#  定义函数，在柱状图上面显示数值
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2.0 -0.2, 1.01*height, '%.2f' % height, size=10, family="Times new roman")


# 幸存人数和死亡人数对比
x = data_train.Survived.value_counts()
plt.subplot(11, 2, 1)
plt.title('幸存死亡人数对比', fontsize=12)
plt.xlim(0, 3)
plt.xticks(np.array(x.index)+1, ['死亡','幸存'])
plt.ylabel('人数')
pb = plt.bar([1,2], np.array(x),  width=0.5, color='dodgerblue')
autolabel(pb)


# 不同等级人数
plt.subplot(11,2,3)
plt.title('不同等级人数', fontsize=12)
plt.ylabel('人数')
plt.xticks([1, 2, 3])
Pclass_counts = data_train.Pclass.value_counts()
pb = plt.bar(Pclass_counts.index, np.array(Pclass_counts), width=0.5, color='orangered')
autolabel(pb)


# 不同等级乘客死亡率
plt.subplot(11, 2, 4)
plt.title('不同等级乘客死亡率', fontsize=12)
Survived_0 = pd.DataFrame(data_train.Pclass[data_train.Survived == 0].value_counts())
Pclass_counts = pd.DataFrame(data_train.Pclass.value_counts())
Survived_0['id'] = pd.Series(Survived_0.index, index=Survived_0.index)
Pclass_counts['id'] = pd.Series(Pclass_counts.index, index=Pclass_counts.index)
data = pd.merge(Survived_0, Pclass_counts, on='id')
plt.xticks([1, 2, 3])
plt.ylabel('死亡率')
pb = plt.bar(data['id'],
        np.array(data['Pclass_x'], dtype=float) / np.array(data['Pclass_y'], dtype=float),
        width=0.5, color='orangered')
autolabel(pb)


# 不同性别人数
plt.subplot(11, 2, 5)
plt.title('不同性别人数',fontsize=12)
Sex = data_train.Sex.value_counts()
plt.xlim(0, 3)
plt.xticks([1, 2],['male', 'fmale'])
plt.ylabel('人数')
pb = plt.bar([1, 2],Sex, width=0.5, color='limegreen')
autolabel(pb)


# 不同性别人数死亡率
plt.subplot(11, 2, 6)
plt.title('不同性别人数死亡率', fontsize=12)
Survived_0 = pd.DataFrame(data_train.Sex[data_train.Survived == 0].value_counts())
plt.xlim(0, 3)
plt.xticks([1, 2],['male', 'fmale'])
plt.ylabel('死亡率')
pb = plt.bar([1, 2], 
        np.array(Survived_0, dtype=float).reshape(-1) / np.array(Sex, dtype=float),
        width=0.5, color='limegreen')
autolabel(pb)


# 不同年龄人数分布
gs = plt.GridSpec(11, 2)
plt.subplot(gs[3, :2])
plt.title('不同年龄人数分布', fontsize=12)
Age = data_train.Age.value_counts()
plt.ylabel('人数')
plt.bar(Age.index, Age, width=0.5)


# 不同年龄段死亡率
# 年龄有缺失，但是大部分的存在，忽略缺失的人数
gs = plt.GridSpec(11,2)
plt.subplot(gs[4, :2])
plt.title('不同年龄段死亡率', fontsize=12)
Survived_0 = pd.DataFrame(data_train.Age[data_train.Survived == 0].value_counts())
Age = pd.DataFrame(Age)
Survived_0['id'] = pd.Series(Survived_0.index, index=Survived_0.index)
Age['id'] = pd.Series(Age.index, index=Age.index)
data = pd.merge(Survived_0, Age, on='id')
plt.ylabel('死亡率')
for i in range(10):
    plt.bar(f'{i * 10}-{(i+1) * 10 - 1}', 
            np.array(data[data['id'] < ((i+1)*10)][data['id'] >= (i *10)]['Age_x'], dtype=float).sum() 
            / np.array(data[data['id'] < ((i+1)*10)][data['id'] >= (i *10)]['Age_y'], dtype=float).sum())
    
# 不同堂兄妹人数分布
plt.subplot(11, 2, 11)
plt.title('堂兄妹人数分布', fontsize=12)
SibSp = data_train.SibSp.value_counts()
SibSp = SibSp.sort_index()
plt.ylabel('人数')
pb = plt.bar(SibSp.index, SibSp, width=0.5, color='violet')
autolabel(pb)


# 不同堂兄妹人数的死亡率
plt.subplot(11, 2, 12)
plt.title('不同堂兄妹人数的死亡率', fontsize=12)
Survived_0 = pd.DataFrame(data_train.SibSp[data_train.Survived == 0].value_counts())
SibSp = pd.DataFrame(SibSp)
Survived_0['id'] = pd.Series(Survived_0.index, index=Survived_0.index)
SibSp['id'] = pd.Series(SibSp.index, index=SibSp.index)
data = pd.merge(Survived_0, SibSp, on='id')
plt.ylabel('死亡率')
pb = plt.bar(data['id'], 
        np.array(data['SibSp_x'], dtype=float) / np.array(data['SibSp_y'], dtype=float),
        width=0.5, color='violet')
autolabel(pb)


# 不同父母小孩人数分布
plt.subplot(11, 2, 13)
plt.title('不同父母小孩人数分布', fontsize=12)
Parch = data_train.Parch.value_counts()
Parch = Parch.sort_index()
plt.ylabel('人数')
pb = plt.bar(Parch.index, Parch, width=0.5, color='violet')
autolabel(pb)


# 不同父母小孩人数的死亡率
plt.subplot(11, 2, 14)
plt.title('不同父母小孩人数的死亡率', fontsize=12)
Survived_0 = pd.DataFrame(data_train.Parch[data_train.Survived == 0].value_counts())
Parch = pd.DataFrame(Parch)
Survived_0['id'] = pd.Series(Survived_0.index, index=Survived_0.index)
Parch['id'] = pd.Series(Parch.index, index=Parch.index)
data = pd.merge(Survived_0, Parch, on='id')
plt.ylabel('死亡率')
pb = plt.bar(data['id'], 
        np.array(data['Parch_x'], dtype=float) / np.array(data['Parch_y'], dtype=float),
        width=0.5, color='violet')
autolabel(pb)


# 不同票价的人数分布
gs = plt.GridSpec(11, 2)
plt.subplot(gs[7, :2])
plt.title('不同票价的人数分布', fontsize=12)
Fare = data_train.Fare.value_counts()
Fare = pd.DataFrame(Fare.sort_index())
Fare['id'] = pd.Series(Fare.index, index=Fare.index)
plt.ylabel('人数')

for i in range(12):
    pb = plt.bar(f'{i * 50}-{(i+1) * 50 - 1}', 
            np.array(Fare[Fare['id'] < ((i+1)*50)][Fare['id'] >= (i *50)]['Fare'], dtype=float).sum(),
            width=0.5)
    autolabel(pb)
    
    
# 不同票价的死亡率
gs = plt.GridSpec(11, 2)
plt.subplot(gs[8, :2])
plt.title('不同票价的死亡率', fontsize=12)
Survived_0 = pd.DataFrame(data_train.Fare[data_train.Survived == 0].value_counts())
Survived_0['id'] = pd.Series(Survived_0.index, index=Survived_0.index)
data = pd.merge(Survived_0, Fare, on='id')
data.fillna(0, inplace=True)
plt.ylabel('死亡率')
for i in range(12):
    plt.bar(f'{i * 50}-{(i + 1) * 50 - 1}', 
             np.array(data[data['id'] < ((i+1)*50)][data['id'] >= (i *50)]['Fare_x'], dtype=float).sum() 
             / np.array(data[data['id'] < ((i+1)*50)][data['id'] >= (i *50)]['Fare_y'], dtype=float).sum())

    
# 登船港口人数分布
plt.subplot(11, 2, 19)
plt.title('登船港口人数分布', fontsize=12)
Embarked = data_train.Embarked.value_counts()
plt.ylabel('人数')
pb = plt.bar(Embarked.index, Embarked, width=0.5)
autolabel(pb)


# 不同登船港口的死亡率
plt.subplot(11, 2, 20)
plt.title('不同登船港口的死亡率', fontsize=12)
Survived_0 = pd.DataFrame(data_train.Embarked[data_train.Survived == 0].value_counts())
Embarked = pd.DataFrame(Embarked)
Embarked['id'] = pd.Series(Embarked.index, index=Survived_0.index)
Survived_0['id'] = pd.Series(Survived_0.index, index=Survived_0.index)
data = pd.merge(Survived_0, Embarked, on='id')
pb = plt.bar(data.id, data['Embarked_x'] / data['Embarked_y'], 
        width=0.5)
plt.ylabel('死亡率')
autolabel(pb)


# 该字段不全，而且缺失值较多，所以按照有无Cabin进行分析
# 有无Cabin的人数分析
plt.subplot(11, 2, 21)
plt.title('有无Cabin的人数', fontsize=12)
Cabin_count = data_train.Cabin.count()
Pid_count = data_train.PassengerId.count()
plt.ylabel('人数')
pb = plt.bar(['有', '无'], [Cabin_count, Pid_count - Cabin_count], width=0.5)
autolabel(pb)


# 有无Cabin的死亡率
plt.subplot(11, 2, 22)
plt.title('有无Cabin的死亡率', fontsize=12)
Survived_cabin = data_train.Survived[pd.notnull(data_train.Cabin)].value_counts()
Survived_nocabin = data_train.Survived[pd.isnull(data_train.Cabin)].value_counts()
plt.ylabel('死亡率')
plt.bar(['有', '无'], 
        [Survived_cabin[0] / (Survived_cabin[1] + Survived_cabin[0]), Survived_nocabin[0] / (Survived_nocabin[0] + Survived_nocabin[1])],
        width=0.5)
```
